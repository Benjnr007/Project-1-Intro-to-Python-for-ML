### Question 18: Which year had the least correlation with â€˜Element Codeâ€™?

```python
print("2014:", df['Element Code'].corr(df['Y2014']))
print("2015:", df['Element Code'].corr(df['Y2015']))
print("2016:", df['Element Code'].corr(df['Y2016']))
print("2017:", df['Element Code'].corr(df['Y2017']))
print("2018:", df['Element Code'].corr(df['Y2018']))
```
